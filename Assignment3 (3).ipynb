```python
# import libraries
import pandas as pd
from sqlalchemy import create_engine
import pymysql
```

```python
# create database connection
engine=create_engine('mysql+pymysql://root:@localhost/data1202')
conn=engine.connect()
```

```python
# Get vgsales data
df=pd.read_sql("select * from vgsales",conn)
```

```python
# calculate average global sales before 2005
avg_sales_before_2005=df[df["Year"]<2005]["Global_Sales"].mean()
```

```python
# calculate average global sales after 2005
avg_sales_after_2005=df[df["Year"]>2005]["Global_Sales"].mean()
```

```python
# compare average global sales
if (avg_sales_before_2005>avg_sales_after_2005):
    print("==> Average Global Sales before 2005 is higher <==")
else:
    print("==> Average Global Sales after 2005 is higher <==")
```

```python
# create a column that labels records before 2005 as 'pre-2005' 
df.loc[df["Year"]<2005,"2005_Label"]="Pre-2005"
```

```python
# create a column that labels records after 2005 as 'post-2005' 
df.loc[df["Year"]>2005,"2005_Label"]="Post-2005"
```
